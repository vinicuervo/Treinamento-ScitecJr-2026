```python
import pandas as pd
```

```python
df = pd.read_csv('Energy_consumption_dataset.csv')
df
```

```python
df.info()
```

```python
df.memory_usage(deep=True)
```

```python
df['DayOfWeek'].unique()
```

```python
df['DayOfWeek'] = df['DayOfWeek'].map({'Monday': 1, 'Tuesday': 2, 'Wednesday': 3, 'Thursday': 4, 'Friday': 5, 'Saturday': 6, 'Sunday': 7}).astype('category')
```

```python
df.memory_usage(deep=True)
```

```python
df['Holiday'].unique()
```

```python
df['Holiday'] = df['Holiday'].map({'No': 0, 'Yes': 1}).astype('category')
```

```python
df.info()
```

```python
df.memory_usage(deep=True)
```

```python
df['HVACUsage'].unique()
```

```python
df['HVACUsage'] = df['HVACUsage'].map({'On': 1, 'Off': 0}).astype('category')
```

```python
df.info()
```

```python
df['HVACUsage'].unique()
```

```python
df.memory_usage(deep=True)
```

```python
df['LightingUsage'].unique()
```

```python
df['LightingUsage'] = df['LightingUsage'].map({'On': 1, 'Off': 0}).astype('category')
```

```python
df['HVACUsage'].unique()
```

```python
df.memory_usage(deep=True)
```

```python
df
```

```python
df.info()
```

```python
df['Month'] = df['Month'].astype('int8')
df['Hour'] = df['Hour'].astype('int8')
df['Occupancy'] = df['Occupancy'].astype('int8')
```

```python
df.memory_usage(deep=True)
```

```python
df['Temperature'] = df['Temperature'].astype('float32')
df['Humidity'] = df['Humidity'].astype('float32')
df['SquareFootage'] = df['SquareFootage'].astype('float32')
df['RenewableEnergy'] = df['RenewableEnergy'].astype('float32')
df['EnergyConsumption'] = df['EnergyConsumption'].astype('float32')
```

```python
df.memory_usage(deep=True)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.to_csv('Energy_consumption_dataset_final.csv', index=False)
```
