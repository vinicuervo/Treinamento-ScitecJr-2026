```python
valores_brutos = [10, 50, 150, 20, 300]

def verificar_sistemas(dados: list[int]):
    alertas = []

    for valor in dados:
        if valor > 100:
            alertas.append(valor)

        if not alertas:
            print('Sistema Seguro')
        else:
            print(f'Revisar valores: {alertas}')

verificar_sistemas(valores_brutos)

valores_brutos = [10, 50, 150, 20, 300]
```
